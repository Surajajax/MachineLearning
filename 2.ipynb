```python
def cost_model(X,y,w,b):
    
    n=X.shape[0]
    cost=0.0
    
    for i in range(n):
        pred_y=np.dot(X[i],w) + b
        cost = cost + (pred_y - y[i])**2
    t_cost=cost / (2* n)
    return float(t_cost)
```

```python
def compute_gradient(X,y,w,b):
    
    n,m=X.shape
    dj_dw=np.zeros((m,))
    dj_db=0.
    
    for i in range(n):
        err = (np.dot(X[i], w) + b) - y[i]   
        for j in range(m):                         
            dj_dw[j] = dj_dw[j] + err * X[i, j]    
        dj_db = dj_db + err                        
    dj_dwe = dj_dw / n                               
    dj_dbe = dj_db / n                              
        
    return dj_dwe, dj_dbe
```

```python
import math
import copy
```

```python
def gradient_descent(X, y, w_in, b_in,cost_function, gradient_function, alpha, num_iters):
    J_history = []
    w = copy.deepcopy(w_in)
    b = b_in
    
    for i in range(num_iters):
        dj_dw,dj_db = gradient_function(X, y, w, b)
        
        w = w - alpha * dj_dw
        b = b - alpha * dj_db
        
        if i<100000:
            J_history.append( cost_function(X, y, w, b))
            
        if i% math.ceil(num_iters / 10) == 0:
            print(f"Iteration {i:4d}: Cost {float(J_history[-1]):8.2f}")

    return w, b, J_history
```

```python
x_scaled = (x_array - x_array.mean(axis=0)) / (x_array.std(axis=0, ddof=0) + 1e-12)
y_scaled = (y_array - y_array.mean()) / (y_array.std() + 1e-12)
y_scaled = y_scaled.ravel()   #


z=x_scaled.shape[1]
w_init=np.zeros(z)
b_init=0.

iterations = 10000
alpha = 0.01




w_final, b_final, J_hist = gradient_descent(x_scaled, y_scaled,w_init, b_init,cost_model,compute_gradient,alpha,iterations)

print(f"b,w found by gradient descent: {b_final:0.2f},{w_final} ")
```

```python
import numpy as np

preds_scaled = x_scaled.dot(w_final) + b_final
m = y_scaled.shape[0]
mse_scaled = np.mean((preds_scaled - y_scaled)**2)
rmse_scaled = np.sqrt(mse_scaled)
ss_res = np.sum((y_scaled - preds_scaled)**2)
ss_tot = np.sum((y_scaled - np.mean(y_scaled))**2)
r2 = 1 - ss_res / ss_tot

print(f"RMSE (scaled): {rmse_scaled:.6f}")
print(f"R^2 (scaled): {r2:.6f}")
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# -----------------------
# 1) PREPARE DATA
# -----------------------
X = x_array                      # (4898, 6)
y = y_array.ravel()              # Make target 1-D

# -----------------------
# 2) OPTIONAL: Scaling (recommended for GD, optional for sklearn LR)
# -----------------------
from sklearn.preprocessing import StandardScaler

scaler_X = StandardScaler()
scaler_y = StandardScaler()

X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y.reshape(-1,1)).ravel()

# -----------------------
# 3) TRAIN SKLEARN LINEAR REGRESSION
# -----------------------
model = LinearRegression()
model.fit(X_scaled, y_scaled)

# -----------------------
# 4) PREDICT
# -----------------------
y_pred_scaled = model.predict(X_scaled)

r2 = r2_score(y_scaled, y_pred_scaled)
rmse = np.sqrt(mean_squared_error(y_scaled, y_pred_scaled))

print(f"RÂ² (scaled): {r2:.6f}")
print(f"RMSE (scaled): {rmse:.6f}")

y_pred_orig = scaler_y.inverse_transform(y_pred_scaled.reshape(-1,1)).ravel()

rmse_orig = np.sqrt(mean_squared_error(y, y_pred_orig))
print(f"RMSE (original units): {rmse_orig:.6f}")
```
